```python
##公鸡 5 母鸡 3 小鸡1/3  100金币 买了 100只鸡
##三重循环
for x in range(0,21):
    for y in range(0,34):
        for z in range(0,301):
            if x+y+z==100 and 5*x+3*y+z/3==100:
                print(x,y,z)
```

```python
##两重循环
## z = 100-(x+y)
## 5*x + 3*y + (100-(x+y))/3 = 100 ##金币
for x in range(0,21):
    for y in range(0,34):
        if 5*x + 3*y + (100-(x+y))/3 == 100:
            print(x,y,100-x-y)
```
