```python
import re
```

```python
target_str = 'xiaoming123'
target_str2 = 'xiaohongxhaoming123'
re.match('xiao',target_str) ##(.*?)代替了x
```

```python
target_str = 'xiaoming123'
target_str2 = 'xiaohongxhiaoming123'
re.match('(.*?)iao',target_str) ##(.*?)代替了x  (.*？)作为一个通配符，就它可以代替任何字符
```

```python
re.findall('(.*?)iao',target_str) ##代替掉了哪个字符
```

```python
re.findall('(.*?)iao',target_str2)  ## 现在需要让她把x和h都找到  re.fin
```

```python
re.findall## 来找(.*?) 代替掉了哪些字符
```
